```python
# Сводка по лабе


# Категориальные признаки — это данные, которые представляют собой категории, а не числа. Например, это может быть цвет машины (красный, синий, зелёный) или страна (Россия, США). Эти признаки нужно преобразовать в числа, чтобы использовать их в моделях машинного обучения.
# 
# Есть несколько способов преобразования:
# 
# One-hot encoding: каждый возможный вариант становится отдельным столбцом с 0 или 1 (например, для цвета: "красный" = [1, 0, 0], "синий" = [0, 1, 0]).
# 
# Label encoding: каждый вариант превращается в число (например, "красный" = 1, "синий" = 2).
# 
# Target encoding: категорию заменяют на среднее значение целевой переменной для этой категории.
# 
# Это основные способы работы с категориальными признаками.
```

```python
# 1. Постройте визуализацию распределения признаков и совместного распределения признаков и целевой переменной с помощью тепловых карт. Попробуйте использовать другие инструменты визуализации.
# Импортируем библиотеки для визуализации
import seaborn as sns  # Для построения красивых графиков и тепловых карт
import matplotlib.pyplot as plt  # Для отображения графиков

# Тепловая карта всех числовых признаков

# Оставляем только числовые столбцы из датасета, так как метод corr работает только с числами
numeric_data = data.select_dtypes(include='number')

# Вычисляем матрицу корреляций между всеми числовыми признаками
corr = numeric_data.corr()

# Создаем фигуру нужного размера
plt.figure(figsize=(12, 8))

# Строим тепловую карту:
# - annot=True — отображаем числовые значения в ячейках
# - cmap='coolwarm' — цветовая палитра (синий — отрицательная связь, красный — положительная)
# - fmt='.2f' — округляем значения до двух знаков
# - linewidths=0.5 — добавляем линии между ячейками
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Добавляем заголовок и отображаем график
plt.title('Тепловая карта корреляций между числовыми признаками')
plt.show()

# Тепловая карта CSAT Score и категориальных признаков

# Выбираем нужные признаки: целевая переменная и категориальные признаки, с которыми хотим найти связь
selected_cols = ['CSAT Score', 'Product_category', 'Agent Shift', 'Tenure Bucket']

# Преобразуем категориальные признаки в числовые с помощью one-hot-кодировки
# drop_first=True исключает первую категорию, чтобы избежать избыточности (ловушка дамми-переменных)
selected_data = pd.get_dummies(data[selected_cols], drop_first=True)

# Создаем фигуру
plt.figure(figsize=(12, 6))

# Строим тепловую карту корреляций между целевой переменной и закодированными признаками
sns.heatmap(selected_data.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Заголовок и отображение
plt.title('Тепловая карта: CSAT Score и категориальные признаки (закодированные)')
plt.show()




#Чем ближе цвет к красному, тем сильнее положительная связь.
#Чем ближе к синему, тем сильнее отрицательная связь. (Отрицательная корреляция означает, что по мере увеличения одного признака — другой уменьшается)
#Белые/нейтральные тона — слабая или нулевая связь.
#
```

```python
# # 2. Построение модели дерева решений и анализ важности признаков + 3. Разбиение датасета на обучающую и тестовую выборки, преобразование данных.
```

```python
# -------------------- Импорт необходимых библиотек --------------------

# Импортируем необходимые библиотеки для работы с моделью дерева решений
from sklearn.tree import DecisionTreeRegressor  # Модель дерева решений для регрессии
from sklearn.model_selection import train_test_split  # Функция для разделения данных на обучающую и тестовую выборки
from sklearn.metrics import mean_squared_error, r2_score  # Метрики для оценки качества модели (среднеквадратичная ошибка и R²)
from sklearn.preprocessing import LabelEncoder  # Класс для кодирования категориальных признаков
import pandas as pd  # Для работы с таблицами (DataFrame)
import numpy as np  # Для работы с массивами и матрицами
import seaborn as sns  # Для визуализации данных с помощью графиков
import matplotlib.pyplot as plt  # Для отображения графиков

# 1. Подготовка данных

# X — это все признаки (все колонки, кроме целевой переменной CSAT Score)
# y — это целевая переменная, которая в нашем случае называется 'CSAT Score' (оценка удовлетворенности)
X = data.drop('CSAT Score', axis=1)  # Удаляем колонку 'CSAT Score' из признаков
y = data['CSAT Score']  # Целевая переменная остается 'CSAT Score'

# Создаем копию X для дальнейшего кодирования категориальных признаков
X_encoded = X.copy()  # Чтобы не изменять исходный DataFrame, создаем его копию

# Словарь для хранения энкодеров (на случай, если потом нужно будет преобразовывать тестовые данные аналогично)
label_encoders = {}

# Проходим по всем столбцам, которые имеют категориальные данные (строки) и преобразуем их в числовые
for col in X_encoded.select_dtypes(include='object').columns:  # Для каждого столбца с типом данных 'object' (строки)
    le = LabelEncoder()  # Создаем объект LabelEncoder для кодирования
    try:
        # Применяем LabelEncoder: он заменяет строковые значения на числа
        X_encoded[col] = le.fit_transform(X_encoded[col].astype(str))  # Преобразуем строковые данные в числовые
        label_encoders[col] = le  # Сохраняем энкодер в словарь для возможного использования в будущем
    except:
        # Если по какой-то причине не получилось закодировать (например, столбец не подходит), удаляем этот столбец
        X_encoded.drop(columns=[col], inplace=True)  # Удаляем невалидный столбец

# 2. Разделение на обучение и тест

# Разделяем данные на обучающую и тестовую выборки
# Обучающая выборка (70% данных) и тестовая выборка (30% данных)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42)

# Это разделение нужно для того, чтобы обучить модель на одной части данных (обучающая выборка),
# а потом проверить её точность на другой части данных (тестовая выборка).

#  3. Обучение модели дерева решений

# Создаем модель дерева решений (это модель регрессии, так как CSAT Score — числовая переменная)
model = DecisionTreeRegressor(random_state=42)  # random_state задает фиксированное случайное состояние для воспроизводимости

# Обучаем модель на обучающих данных
model.fit(X_train, y_train)  # Мы передаем обучающие данные и целевые значения для обучения

#4. Предсказание и оценка

# Модель теперь обучена, и мы можем использовать её для предсказания значений на тестовой выборке
y_pred = model.predict(X_test)  # Делаем предсказания по тестовым данным

# Оценка точности модели:
# Мы используем среднеквадратичную ошибку (MSE) и коэффициент детерминации R² для оценки модели.
mse = mean_squared_error(y_test, y_pred)  # MSE — это средняя квадратная ошибка между реальными и предсказанными значениями
r2 = r2_score(y_test, y_pred)  # R² — это коэффициент детерминации, который показывает, насколько хорошо модель объясняет данные

# Выводим метрики
print(f"Среднеквадратичная ошибка (MSE): {mse:.2f}")  # Среднеквадратичная ошибка
print(f"Коэффициент детерминации (R²): {r2:.2f}")  # Коэффициент детерминации R²

#5. Анализ важности признаков

# Теперь, когда модель обучена, мы можем узнать, какие признаки (колонки) были наиболее важными для предсказания.
importances = model.feature_importances_  # Это массив важности признаков для дерева решений

# Определим, сколько признаков мы будем показывать на графике (например, топ-20 самых важных)
top_n = 20  # Покажем топ-20 признаков по важности
indices = np.argsort(importances)[-top_n:]  # Индексы признаков, отсортированные по убыванию важности

#6. Визуализация

# Строим график важности признаков
plt.figure(figsize=(10, 6))  # Устанавливаем размер графика

# Строим горизонтальный столбчатый график: на оси X — важность, на оси Y — названия признаков
sns.barplot(x=importances[indices], y=X_encoded.columns[indices])

# Добавляем заголовок и подписи к осям
plt.title('ТОП-20 важных признаков для модели дерева решений')
plt.xlabel('Важность признака')  # Ось X — важность признаков
plt.ylabel('Название признака')  # Ось Y — названия признаков

# Отображаем график
plt.tight_layout()  # Убираем лишние отступы для улучшенного отображения
plt.show()
```
