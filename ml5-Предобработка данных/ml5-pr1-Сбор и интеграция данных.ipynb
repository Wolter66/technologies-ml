```python
# Сводка по лабе

# В этой лабораторной работе мы объединили данные из нескольких источников, чтобы создать единый набор информации, который можно анализировать. Начали с того, что загрузили три файла с данными о клиентских обращениях в службу поддержки. Эти файлы содержали похожую информацию, но немного отличались по структуре. Поэтому сначала мы внимательно изучили их, чтобы понять, как они устроены, и привели их к одинаковому виду. Это важно, потому что для объединения данных они должны быть совместимыми, например, иметь одинаковые названия столбцов.
# (Чтобы привести данные к нормальному виду, мы проверили файлы на пропуски и лишние значения, удалили или заменили их. Привели даты и числовые значения к единому формату. Согласовали названия столбцов, чтобы они совпадали во всех файлах.)
# 
# После этого мы объединили эти три файла в один с помощью вертикальной интеграции. Это значит, что мы просто добавили строки из одного файла к строкам другого, создав единый список всех обращений. Такой подход позволяет нам собрать всю информацию в одном месте и упорядочить её.
# 
# Затем мы добавили информацию о заказах клиентов, чтобы привязать её к обращениям в службу поддержки. Это нужно для того, чтобы можно было анализировать, как заказы связаны с обращениями клиентов, и как часто клиенты обращаются за помощью.
# 
# Далее мы подключили данные о менеджерах, которые работают в службе поддержки, чтобы дополнительно учитывать информацию о том, какие сотрудники обслуживали какие обращения. Это важно для анализа производительности сотрудников и понимания, как они влияли на решение проблем клиентов.
# 
# В итоге мы создали единый датасет, в котором содержится вся нужная информация: обращения клиентов, данные о заказах и информация о менеджера
```

```python
# 1. Убедиться в отсутствии дубликатов после вертикального объединения
#    - Сначала проверить уникальность ключа (Id) в каждом датасете
print("Unique Ids in data1:", data1['Id'].is_unique)
print("Unique Ids in data2:", data2['Id'].is_unique)
print("Unique Ids in data3:", data3['Id'].is_unique)

#    - Объединить и проверить дубликаты по Id:
combined_12 = pd.concat([data1, data2.rename(columns={'Unique id': 'Id'})], ignore_index=True)
duplicates = combined_12[combined_12.duplicated(subset=['Id'], keep=False)]
print("Number of duplicates after concat data1 and data2:", duplicates.shape[0])
```

```python
from matplotlib import pyplot as plt

merged = data1.merge(data2, on='Order_id', how='inner')

#    Построим гистограммы / плотности для числовых признаков:
numeric_cols = ['CSAT Score', 'Channel']  # замените списком нужных
for col in numeric_cols:
    plt.figure()
    sns.histplot(data1[col].dropna(), kde=True)
    plt.title(f"Distribution of {col} in data1")

    plt.figure()
    sns.histplot(data2[col].dropna(), kde=True)
    plt.title(f"Distribution of {col} in data2")
```

```python
# Настройка отображения графиков
plt.rcParams['figure.figsize'] = (8, 5)

# 1. Вертикальное объединение: проверка дубликатов
# ------------------------------------------------
# Вертикальная конкатенация
data_all = pd.concat([data1, data2, data3], ignore_index=True)

# Проверка дубликатов по уникальному ключу обращения (Id)
dup_count = data_all.duplicated(subset=['Id']).sum()
print(f"Количество дубликатов по Id после вертикального объединения: {dup_count}")
if dup_count > 0:
    duplicates = data_all[data_all.duplicated(subset=['Id'], keep=False)]
    print("Список дублирующихся записей:")
    print(duplicates[['Id', 'Channel', 'Issue_reported_Date']].head(10))

# 2. Горизонтальная интеграция data1 и data2: визуализация распределений
# -----------------------------------------------------------------------
# Подготовка: переименование и приведение к datetime
data1_clean = data1.copy()
data2_clean = data2.copy()
data2_clean.rename(columns={
    'channel_name': 'Channel',
    'Reported at': 'Issue_reported_Date',
    'Responded at': 'Issue_responded_Date',
    'Survey responded at': 'Survey_response_Date'
}, inplace=True)

for df in (data1_clean, data2_clean):
    for col in ['Issue_reported_Date', 'Issue_responded_Date', 'Survey_response_Date']:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Inner join по Order_id
merged_12 = data1_clean.merge(data2_clean, on='Order_id', how='inner', suffixes=('_1', '_2'))
print(f"Размер объединённой таблицы data1 и data2 (inner): {merged_12.shape}")

# Автоматическое получение всех числовых столбцов для визуализации
numeric_cols = merged_12.select_dtypes(include=[np.number]).columns
print(f"Числовые столбцы для визуализации распределений: {list(numeric_cols)}")

# Вывод гистограмм и KDE для каждого числового столбца
for col in numeric_cols:
    plt.figure()
    sns.histplot(merged_12[col].dropna(), kde=True)
    plt.title(f"Распределение {col} в объединённой таблице data1+data2")
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

# 3. Верификация пропусков после объединения с orders.csv
# ------------------------------------------------------
data_with_orders = data_all.merge(orders_data, on='Order_id', how='left')

# Процент пропусков по столбцам
missing_stats = data_with_orders.isnull().mean() * 100
print("Процент пропусков в каждом столбце после объединения с orders.csv:")
print(missing_stats.sort_values(ascending=False))

# Точечные проверки первых 5 Order_id
sample_ids = data_with_orders['Order_id'].dropna().unique()[:5]
print("\nТочечная проверка первых 5 Order_id:")
for oid in sample_ids:
    print(f"\nOrder_id: {oid}")
    print("В таблице orders:")
    print(orders_data[orders_data.Order_id == oid])
    print("В объединённой таблице:")
    display(data_with_orders[data_with_orders.Order_id == oid].iloc[:2][['Id','Order_id','order_date_time']])

# Случайная проверка 3 Order_id
random_ids = np.random.choice(sample_ids, size=3, replace=False)
print("\nСлучайная проверка 3 Order_id:")
for oid in random_ids:
    exists = not orders_data[orders_data.Order_id == oid].empty
    print(f"Order_id: {oid}, есть в orders: {exists}")

print("\nПроверки выполнены успешно.")
```
