```python
from sklearn.datasets import make_classification
```

```python

X, y = make_classification(n_samples=1000, n_classes=2, random_state=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression().fit(X_train, y_train)
```

```python
y_pred_proba = model.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, _ = roc_curve(y_test, y_pred_proba[:, 1])
```

```python
from matplotlib import pyplot as plt

plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred_proba[:, 1])
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, _ = precision_recall_curve(y_test, y_pred_proba[:, 1])
```

```python
no_skill = len(y[y==1]) / len(y)
plt.plot([0, 1], [no_skill, no_skill], linestyle='--')
plt.plot(recall, precision, marker='.')
plt.show()
```

```python
from sklearn.metrics import auc

auc(recall, precision)
```

```python
from sklearn.metrics import classification_report

y_test_pred = model.predict(X_test)
print(classification_report(y_test, y_test_pred))
```

```python
# Повторите анализ из лабораторной работы, но с двумерными датасетами. Изобразите графически, как изменение порога влияет на расположение границы принятия решений.
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np

# Создаём двумерный датасет
X, y = make_classification(n_samples=500, n_features=2, n_redundant=0, n_informative=2,
                           n_clusters_per_class=1, class_sep=1.5, random_state=42)

model = LogisticRegression().fit(X, y)
proba = model.predict_proba(X)[:, 1]

# Список порогов для визуализации
thresholds = [0.9, 0.7, 0.5, 0.3, 0.1]

xx, yy = np.meshgrid(np.linspace(X[:,0].min()-1, X[:,0].max()+1, 200),
                     np.linspace(X[:,1].min()-1, X[:,1].max()+1, 200))

Z_input = np.c_[xx.ravel(), yy.ravel()]
Z_proba = model.predict_proba(Z_input)[:, 1]

plt.figure(figsize=(15, 10))

for i, thresh in enumerate(thresholds):
    plt.subplot(2, 3, i+1)
    Z = (Z_proba > thresh).astype(int).reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', edgecolors='k')
    plt.title(f"Порог: {thresh:.2f}")
    plt.xlabel("Признак 1")
    plt.ylabel("Признак 2")

plt.tight_layout()
plt.show()
```

```python
# Повторите анализ на реальном датасете для бинарной классификации. Проинтерпретируйте результат, сделайте вывод.
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, precision_recall_curve, auc
import matplotlib.pyplot as plt

# Загрузка датасета
data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

model = LogisticRegression(max_iter=10000).fit(X_train, y_train)
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

# Метрики
print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))

# Precision-Recall
precision, recall, _ = precision_recall_curve(y_test, y_proba)
pr_auc = auc(recall, precision)
print("PR AUC:", pr_auc)

# График PR-кривой
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(recall, precision, marker='.')
plt.title("Precision-Recall кривая")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.grid(True)
plt.show()
```

```python
# В задании по оптимизации порога используйте ROC и среднее геометрическое между TPR и FPR как критерий оптимизации.
```

```python
from sklearn.metrics import roc_curve
from numpy import sqrt, argmax

# ROC кривая
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

# Среднее геометрическое между TPR и (1 - FPR)
g = sqrt(tpr * (1 - fpr))
ix = argmax(g)
best_threshold = thresholds[ix]

print(f"Оптимальный порог: {best_threshold:.3f}, Geometric mean: {g[ix]:.3f}")

# Построим график
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label="Best Threshold")
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.title("ROC-кривая с оптимизацией G-Mean")
plt.grid(True)
plt.show()

# Прогноз по оптимальному порогу
y_pred_opt = (y_proba > best_threshold).astype(int)
print(classification_report(y_test, y_pred_opt))
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, classification_report
import numpy as np
import matplotlib.pyplot as plt

# 1) Реальный бинарный датасет
data = load_breast_cancer(as_frame=True)
X, y = data.data, data.target
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3,
                                          random_state=42, stratify=y)

# 2) Обучаем модель и получаем вероятности
model = LogisticRegression(max_iter=5000).fit(X_tr, y_tr)
y_proba = model.predict_proba(X_te)[:, 1]
y_pred_default = model.predict(X_te)

# 3) Базовый отчёт
print("=== Отчёт при пороге 0.5 ===")
print(classification_report(y_te, y_pred_default,
                            target_names=data.target_names))

# 4) ROC + поиск оптимального порога по G-mean
fpr, tpr, thresholds = roc_curve(y_te, y_proba)
g_mean = np.sqrt(tpr * (1 - fpr))
idx_opt = np.argmax(g_mean)
best_thresh = thresholds[idx_opt]
print(f"\nОптимальный порог = {best_thresh:.3f}, G-mean = {g_mean[idx_opt]:.3f}")

# 5) График ROC с отмеченной точкой
plt.figure()
plt.plot(fpr, tpr, marker='.', label='ROC-кривая')
plt.scatter(fpr[idx_opt], tpr[idx_opt], 
            color='red', label=f'Opt thresh = {best_thresh:.2f}')
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel('FPR'); plt.ylabel('TPR')
plt.title('ROC-кривая и оптимальный порог')
plt.legend(); plt.show()

# 6) Отчёт с оптимальным порогом
y_pred_opt = (y_proba >= best_thresh).astype(int)
print("\n=== Отчёт при оптимальном пороге ===")
print(classification_report(y_te, y_pred_opt,
                            target_names=data.target_names))
```
