```python
from sklearn.datasets import make_classification
```

```python

X, y = make_classification(n_samples=1000, n_classes=2, random_state=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression().fit(X_train, y_train)
```

```python
y_pred_proba = model.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, _ = roc_curve(y_test, y_pred_proba[:, 1])
```

```python
from matplotlib import pyplot as plt

plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred_proba[:, 1])
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, _ = precision_recall_curve(y_test, y_pred_proba[:, 1])
```

```python
no_skill = len(y[y==1]) / len(y)
plt.plot([0, 1], [no_skill, no_skill], linestyle='--')
plt.plot(recall, precision, marker='.')
plt.show()
```

```python
from sklearn.metrics import auc

auc(recall, precision)
```

```python
from sklearn.metrics import classification_report

y_test_pred = model.predict(X_test)
print(classification_report(y_test, y_test_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model2 = KNeighborsClassifier(n_neighbors=100).fit(X_train, y_train)
y_pred2_proba = model2.predict_proba(X_test)
y_pred1_proba = model.predict_proba(X_test)
fpr, tpr, _ = roc_curve(y_test, y_pred2_proba[:, 1])
fpr1, tpr1, _ = roc_curve(y_test, y_pred1_proba[:, 1])

plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.plot(fpr1, tpr1, marker='.')
plt.show()
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np

#   
X, y = make_classification(n_samples=500, n_features=2, n_redundant=0, n_informative=2,
                           n_clusters_per_class=1, class_sep=1.5, random_state=42)

model = LogisticRegression().fit(X, y)
proba = model.predict_proba(X)[:, 1]

#    
thresholds = [0.9, 0.7, 0.5, 0.3, 0.1]

xx, yy = np.meshgrid(np.linspace(X[:,0].min()-1, X[:,0].max()+1, 200),
                     np.linspace(X[:,1].min()-1, X[:,1].max()+1, 200))

Z_input = np.c_[xx.ravel(), yy.ravel()]
Z_proba = model.predict_proba(Z_input)[:, 1]

plt.figure(figsize=(15, 10))

for i, thresh in enumerate(thresholds):
    plt.subplot(2, 3, i+1)
    Z = (Z_proba > thresh).astype(int).reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', edgecolors='k')
    plt.title(f": {thresh:.2f}")
    plt.xlabel(" 1")
    plt.ylabel(" 2")

plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, precision_recall_curve, auc
import matplotlib.pyplot as plt

#  
data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

model = LogisticRegression(max_iter=10000).fit(X_train, y_train)
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

# 
print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))

# Precision-Recall
precision, recall, _ = precision_recall_curve(y_test, y_proba)
pr_auc = auc(recall, precision)
print("PR AUC:", pr_auc)

#  PR-
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(recall, precision, marker='.')
plt.title("Precision-Recall ")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.grid(True)
plt.show()
```

```python
#       ROC     TPR  FPR   .
```

```python
from sklearn.metrics import roc_curve
from numpy import sqrt, argmax

# ROC 
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

#    TPR  (1 - FPR)
g = sqrt(tpr * (1 - fpr))
ix = argmax(g)
best_threshold = thresholds[ix]

print(f" : {best_threshold:.3f}, Geometric mean: {g[ix]:.3f}")

#  
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label="Best Threshold")
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.title("ROC-   G-Mean")
plt.grid(True)
plt.show()

#    
y_pred_opt = (y_proba > best_threshold).astype(int)
print(classification_report(y_test, y_pred_opt))
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, classification_report
import numpy as np
import matplotlib.pyplot as plt

# 1)   
data = load_breast_cancer(as_frame=True)
X, y = data.data, data.target
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3,
                                          random_state=42, stratify=y)

# 2)     
model = LogisticRegression(max_iter=5000).fit(X_tr, y_tr)
y_proba = model.predict_proba(X_te)[:, 1]
y_pred_default = model.predict(X_te)

# 3)  
print("===    0.5 ===")
print(classification_report(y_te, y_pred_default,
                            target_names=data.target_names))

# 4) ROC +     G-mean
fpr, tpr, thresholds = roc_curve(y_te, y_proba)
g_mean = np.sqrt(tpr * (1 - fpr))
idx_opt = np.argmax(g_mean)
best_thresh = thresholds[idx_opt]
print(f"\n  = {best_thresh:.3f}, G-mean = {g_mean[idx_opt]:.3f}")

# 5)  ROC   
plt.figure()
plt.plot(fpr, tpr, marker='.', label='ROC-')
plt.scatter(fpr[idx_opt], tpr[idx_opt], 
            color='red', label=f'Opt thresh = {best_thresh:.2f}')
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel('FPR'); plt.ylabel('TPR')
plt.title('ROC-   ')
plt.legend(); plt.show()

# 6)    
y_pred_opt = (y_proba >= best_thresh).astype(int)
print("\n===     ===")
print(classification_report(y_te, y_pred_opt,
                            target_names=data.target_names))
```
