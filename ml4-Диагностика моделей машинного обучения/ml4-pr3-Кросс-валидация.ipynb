```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris

iris = load_iris()
X=iris.data
y=iris.target
```

```python
iris_data = pd.DataFrame(iris['data'], columns=iris['feature_names'])
name_map = {0: 'Iris-setosa', 1: 'Iris-versicolor', 2:'Iris-virginica'}
iris_data['class'] = [name_map[k] for k in iris['target']]
iris_data.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=4)
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train) #Обучение трейновой выборке
y_pred = model.predict(X_test) #Предсказание для тестовой выборки
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred, average='macro'))
```

```python
from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

```python
from sklearn.model_selection import KFold,StratifiedKFold,LeaveOneOut, cross_val_score
```

```python
kf = KFold(n_splits = 3,shuffle=True, random_state=15)
kf
```

```python
for i, (train_index, test_index) in enumerate(kf.split(y)):
    print("Fold {}: Длинна train: {}, Длинна test: {}".format(i+1, len(train_index), len(test_index)))
    print('Train: index={}\n Test:  index={}'.format(train_index, test_index))

    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
metrics_accuracy = []
metrics_f1 = []
model = LogisticRegression(solver='liblinear')
for i, (train_index, test_index) in enumerate(kf.split(y)):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    metrics_accuracy.append(accuracy_score(y_test, y_pred))
    metrics_f1.append(f1_score(y_test, y_pred, average='macro'))
```

```python
print('Значения метрики accuracy: {} \nЗначения метрики f1: {}'.format(metrics_accuracy, metrics_f1))
```

```python
import numpy as np
print("Среднее по кросс-валидации: ", np.array(metrics_f1).mean())
```

```python
cv_results = cross_val_score(model,                  # модель
                             X,                      # матрица признаков
                             y,                      # вектор цели
                             cv = kf,                # тип разбиения (можно указать просто число фолдов cv = 3)
                             scoring = 'accuracy',   # метрика
                             n_jobs=-1)              # используются все ядра CPU

print("Кросс-валидация: ", cv_results)
print("Среднее по кросс-валидации: ", cv_results.mean())
print("Дисперсия по кросс-валидации: ", cv_results.std())
```

```python
skf = StratifiedKFold(n_splits=3,shuffle=True, random_state=15)
skf.get_n_splits(X, y)
```

```python
for i, (train_index, test_index) in enumerate(skf.split(X,y)):
    print(f"Fold {i+1}:")
    print('Train: index={}\n Test:  index={}'.format(train_index, test_index))

    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
cv_results = cross_val_score(model,                  # модель
                             X,                      # матрица признаков
                             y,                      # вектор цели
                             cv = skf,           # тип разбиения
                             scoring = 'f1_macro',   # метрика
                             n_jobs=-1)              # используются все ядра CPU

print("Кросс-валидация: ", cv_results)
print("Среднее по кросс-валидации: ", cv_results.mean())
```

```python
loo = LeaveOneOut()
```

```python
for i, (train_index, test_index) in enumerate(loo.split(X)):
    print(f"Fold {i+1}:")
    print('Train: index={}\n Test:  index={}'.format(train_index, test_index))

    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
cv_results = cross_val_score(model,                  # модель
                             X,                      # матрица признаков
                             y,                      # вектор цели
                             cv = loo,           # тип разбиения
                             scoring = 'f1_macro',   # метрика
                             n_jobs=-1)              # используются все ядра CPU

print("Кросс-валидация: ", cv_results)
print("Среднее по кросс-валидации: ", cv_results.mean())
```

```python
# Сводка по лабе

# В лабораторной работе мы загрузили данные о цветах ирисов и преобразовали их в удобный формат для анализа. Мы разделили эти данные на признаки (например, длина и ширина лепестков) и классы (разные виды ирисов). Чтобы оценить работу модели, мы использовали две основные метрики: точность, которая показывает, сколько правильных предсказаний сделала модель, и F1-меру, которая учитывает как правильность, так и полноту предсказаний.
# 
# Для того чтобы получить более точную оценку работы модели, мы использовали метод, называемый кросс-валидацией. Это способ, при котором данные делятся на несколько частей. Модель обучается на одних частях данных, а проверяется на других, и так несколько раз. Это помогает уменьшить случайные ошибки, которые могут возникать при одном разбиении данных.
# 
# Мы использовали разные подходы к разбиению данных: один из них просто делит данные на несколько частей (классическая кросс-валидация), другой — так, чтобы в каждой части было одинаковое количество представителей каждого класса (стратифицированная кросс-валидация), а третий — когда каждый объект в данных по очереди используется как тест, а все остальные как обучающие (метод leave-one-out). В результате таких проверок мы получили более надежные и точные оценки работы модели.
```
